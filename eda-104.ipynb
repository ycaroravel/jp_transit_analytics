```python
#importar dados de localização dos onibus
passagens1 = pd.read_csv('SJ02 - 21032017 - Linha 1041.csv', delimiter=';')
passagens2 = pd.read_csv('SJ02 - 21032017 - Linha 1042.csv', delimiter=';')
passagens3 = pd.read_csv('SJ02 - 21032017 - Linha 1043.csv', delimiter=';')
passagens = pd.concat([passagens1, passagens2, passagens3]).reset_index(drop=True)
```

```python
passagens = passagens.drop(passagens[(passagens.Parada=='FIM 104A')|(passagens.Parada=='FIM 104B')|(passagens.Parada=='FIM 104C')].index, axis=0).reset_index(drop=True)
```

```python
#formatar dados de localização dos onibus
passagens['DataHoraAntesParada'] =  pd.to_datetime(passagens['DataHoraAntesParada'], format='%d/%m/%Y %H:%M')
passagens['DataHoraDepoisParada'] =  pd.to_datetime(passagens['DataHoraDepoisParada'], format='%d/%m/%Y %H:%M')
passagens['DataHoraMedianaParada'] = passagens['DataHoraAntesParada'] + ((passagens['DataHoraDepoisParada'] - passagens['DataHoraAntesParada'])/2)
passagens = passagens.sort_values('DataHoraAntesParada').reset_index(drop=True)
passagens['NumeroParada'] = pd.to_numeric(passagens.Parada.str.extract('([0-9][0-9])', expand=False))
passagens.head()
```

```python
#separar viagens
passagens['aux'] = 0
passagens['Viagem'] = ""
#analisa as viagens carro por carro
for carro in passagens.Veiculo.unique():
    #coluna aux: diferença enter o valor atual e anterior na coluna numero parada
    passagens.loc[passagens.Veiculo==carro, 'aux'] = passagens[passagens.Veiculo==carro].NumeroParada.diff().abs()
    #coluna aux: quando a diferença é menor que 10, substituir por 0
    #quando a diferença para o valor anterior for pequena, aux será 0
    passagens.loc[(passagens.Veiculo==carro) & ((passagens.aux.isnull()) | (passagens.aux<10)), 'aux'] = 0
    #coluna aux: para todos os outros valores, substituir por 1
    #quando a diferença para o valor anterior for grande (ex.: 68 para 1), aux será 1, sinalizando um degrau
    passagens.loc[(passagens.Veiculo==carro) & (passagens.aux!=0), 'aux'] = 1
    #coluna aux: utilizar a soma cumulativa dos valoles na coluna
    #entradas da primeira viagem terão aux 0, da segunda terão aux 1, da terceira aux 2...
    passagens.loc[passagens.Veiculo==carro, 'aux'] = passagens.aux[passagens.Veiculo==carro].cumsum()
    #concatenar Veiculo e aux para ids únicos para cada viagem
    passagens.loc[passagens.Veiculo==carro, 'Viagem'] = pd.to_numeric(passagens.Veiculo[passagens.Veiculo==carro].astype('str') + passagens.aux[passagens.Veiculo==carro].astype('str'))
```

```python
passagens2 = passagens.copy()
#passagens2 = passagens2[~passagens2.Viagem.isin(passagens2.Viagem.value_counts()[passagens2.Viagem.value_counts().values<10].index)].sort_values('DataHoraAntesParada').reset_index(drop=True)
```

```python
#inserir primeira parada para viagens que não tem esse registro
#analisa cada viagem
for viagem in passagens2.Viagem.unique():
    #resgata a primeira linha e verifica se o NumeroParada é 1, caso não...
    if (passagens2[passagens2.Viagem==viagem].head(1).NumeroParada.values[0] != 1):
        i = passagens2[passagens2.Viagem==viagem].head(1).index[0]
        #copia a primeira linha
        line = passagens2[passagens2.Viagem==viagem].head(1)
        line.Parada = 'P01 I 301'
        line.NumeroParada = 1
        #utliza a hora do último registro da viagem anterior do mesmo carro, quando há
        if ((viagem-1) in passagens2.Viagem.unique()):
            line.DataHoraAntesParada = passagens2[passagens2.Viagem==(viagem-1)].tail(1).DataHoraAntesParada.values[0] + pd.Timedelta('00:00:01')
        #insere a nova linha
        passagens2 = pd.concat([passagens2.iloc[:i], pd.concat([line, passagens2.iloc[i:]])]).reset_index(drop=True)

passagens2 = passagens2.sort_values('DataHoraAntesParada')
```

```python
passagens2['Carro'] = passagens2['Veiculo']
cartoes2 = pd.merge_asof(cartoes, passagens2[['NumeroParada', 'DataHoraAntesParada', 'DataHoraDepoisParada', 'DataHoraMedianaParada', 'Viagem', 'Carro', 'Rota']], by='Carro', right_on='DataHoraAntesParada', left_on='DataHoraAcesso')
```

## Tabelas por tipo e viagem

```python
tipos = ['Pagante-DEB', 'Comum-DEB', 'Funcionário-DEB', 'Gratuito-DEB',
       'Estudante-DEB', 'Comum-INT', 'Estudante-INT']

tables = {}
for tipo in tipos:
    pivoted3 = pd.pivot_table(cartoes3[cartoes3.Tipo==tipo][['NumeroParada', 'Viagem']], index=['NumeroParada'], columns=['Viagem'], aggfunc=lambda x: len(x))
    #a[a.isnull()] = 0
    a = pd.DataFrame(index=list(range(1,69)), columns=cartoes3.Viagem.unique().astype('str'), dtype='float64')
    b = pd.DataFrame(pivoted3.to_records()).set_index('NumeroParada', drop=True)

    table = a.combine_first(b)
    table[table.isnull()] = 0
    table.loc[:, 'Total'] = table.sum(axis=1)
    table.loc['Total', :] = table.sum(axis=0)
    
    tables[tipo] = table
```

```python
tables['Total'] = tables['Pagante-DEB'] + tables['Comum-DEB'] + tables['Funcionário-DEB'] + tables['Gratuito-DEB'] + tables['Estudante-DEB'] + tables['Comum-INT'] + tables['Estudante-INT']
```

```python
viagens = table.columns
infos_viagens = {}
for viagem in viagens:
    a = pd.DataFrame(columns = np.append(tipos, 'Total'))
    for tipo in a.columns:
        a[tipo] = tables[tipo][viagem]
    infos_viagens[viagem] = a
```

<i>tables</i> é um dicionário em que cada chave é um tipo de passagem (ou o 'Total', a soma de todos os tipos). Para acessar as informações de um tipo, utilizar <i>tables['tipo']</i>. O resultado será um DF onde cada coluna representa uma viagem, cada linha uma parada, a última coluna e linha os totais em cada direção e os valores são as entradas de passageiros daqueles tipo.
